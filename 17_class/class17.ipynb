```python
total : int = 10000

s1 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(80,100, (total,5))
ss1 : pd.DataFrame = pd.DataFrame(s1, columns=['sub1','sub2','sub3','sub4','sub5'])

s2 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(70,79, (total,5))
ss2 : pd.DataFrame = pd.DataFrame(s2, columns=['sub1','sub2','sub3','sub4','sub5'])

s3 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(60,69, (total,5))
ss3 : pd.DataFrame = pd.DataFrame(s3, columns=['sub1','sub2','sub3','sub4','sub5'])

s4 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(50,59, (total,5))
ss4 : pd.DataFrame = pd.DataFrame(s4, columns=['sub1','sub2','sub3','sub4','sub5'])

s5 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(40,49, (total,5))
ss5 : pd.DataFrame = pd.DataFrame(s5, columns=['sub1','sub2','sub3','sub4','sub5'])

s6 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(33,39, (total,5))
ss6 : pd.DataFrame = pd.DataFrame(s6, columns=['sub1','sub2','sub3','sub4','sub5'])

s7 : NDArray[Shape[str(total)+"5"], Int32] = np.random.randint(0,32, (total,5))
ss7 : pd.DataFrame = pd.DataFrame(s7, columns=['sub1','sub2','sub3','sub4','sub5'])
```

## apply custom function on columns(SeriesData)

```python
def grade(per:int)->str:
    if per >=80:
        return "A+"
    elif per >=70:
        return "B"
    elif per >=60:
        return "C"
    elif per >=50:
        return "D"
    elif per >=40:
        return "E"
    elif per >=33:
        return "E"
    else:
        return "F"
    

    
```

```python
df.head()
```

```python
df['Percentage'].apply(grade)
```

```python
df['Grade'] = df['Percentage'].apply(grade)
df
```

# map function for remark

```python
from typing import Dict


remark : Dict[str,str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F" : "Fail"
} 

df["Remark"] = df.Grade.map(remark)
df
```
